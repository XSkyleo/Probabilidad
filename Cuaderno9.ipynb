```python
import scipy.stats as stats
import math

import matplotlib.pyplot as plt
import numpy as np

x_bar = 72.7452
mu_0 = 77.8341
sigma = 87.1737
n = 50
alpha = 0.05

z = (x_bar - mu_0) / (sigma / math.sqrt(n))

z_critical = stats.norm.ppf(1 - alpha / 2) # Para dos colas

z_values = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_values)

plt.figure(figsize=(8, 6))
plt.plot(z_values, pdf_values, label='Distribución Normal', color='blue')
plt.fill_between(z_values, pdf_values, where=(z_values <= -z_critical) | (z_values >= z_critical),
color='red', alpha=0.5, label='Área de rechazo')

plt.axvline(z, color='green', linestyle='dashed', label=f'Valor calculado de z: {z:.2f}')
plt.axvline(-z_critical, color='orange', linestyle='dashed', label=f'Valor crítico z: ±{z_critical:.2f}')
plt.axvline(z_critical, color='orange', linestyle='dashed')

# Agregar etiquetas y título
plt.title('Prueba de Hipótesis: Distribución Normal y Área de Rechazo')
plt.xlabel('Valor de z')
plt.ylabel('Densidad de probabilidad')
plt.legend(loc='upper left')

# Mostrar la gráfica
plt.grid(True)
plt.show()

# Mostrar el valor de z
print(f"Valor calculado de z: {z}")
print(f"Valor crítico de z para una prueba bilateral (α = 0.05): {z_critical}")

# Comparar el valor calculado de z con los valores críticos
if abs(z) > z_critical:
  print("Rechazamos la hipótesis nula.")
else:
  print("No rechazamos la hipótesis nula. No hay suficiente evidencia para rechazar que no hay diferencia significativa entre las calificaciones antes y después.")

print("Diferencia de medias")

import matplotlib.pyplot as plt
import numpy as np
categories = ['Antes', 'Después']
frequencies = [72.7452, 77.8341]
plt.bar(categories, frequencies, color='skyblue', alpha=0.7, edgecolor='black')
plt.title('Diagrama de Barras: Diferencia de medias')
plt.xlabel('Tiempo')
plt.ylabel('Calificación')
plt.show()
```
