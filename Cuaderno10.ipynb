```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos del problema
media_muestral = 76.96 # x
media_poblacional = 78.28 # 
desviacion_muestral = 76.11 # s
n = 100 # Tamao de la muestra
alpha = 0.05 # Nivel de significancia

# Paso 2: Clculo del estadstico t
t_calculado = (media_muestral - media_poblacional) / (desviacion_muestral / np.sqrt(n))

# Paso 3: Determinacin del valor crtico
df = n - 1 # Grados de libertad
t_critico = stats.t.ppf(1 - alpha / 2, df) # Valor crtico de t para prueba bilateral

# Paso 4: Grfica de la distribucin t-Student
x = np.linspace(-4, 4, 1000) # Rango de valores en el eje x
y = stats.t.pdf(x, df) # Funcin de densidad de probabilidad

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribucin t-Student", color='blue')

# Sombrar la regin de rechazo
x_rechazo_derecha = np.linspace(t_critico, 4, 100)
y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Regin de rechazo")

x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

# Lneas de referencia
plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"t crtico = {t_critico:.2f}")
plt.axvline(x=-t_critico, color='black', linestyle='dashed')
plt.axvline(x=t_calculado, color='green', linestyle='dashed', label=f"t calculado = {t_calculado:.2f}")

plt.title("Distribucin t-Student y regiones de rechazo")
plt.xlabel("Valor de t")
plt.ylabel("Densidad de probabilidad")
plt.legend()

plt.grid()
plt.show()

# Paso 5: Comparacin y decisin
print(f"Valor calculado de t: {t_calculado:.2f}")
print(f"Valor crtico de t (bilateral, ={alpha}): {t_critico:.3f}")

if abs(t_calculado) > t_critico:
    print("Rechazamos la hiptesis nula. La media del grupo es significativamente diferente.")
else:
    print("No rechazamos la hiptesis nula. No hay evidencia suficiente para afirmar una diferencia.")
```

```python
#Hiptesis nula (H0): Los tiempos de entrega de los lotes A y B son iguales.
#Hiptesis alternativa (H1): Los tiempos de entrega de los lotes A y B son diferentes.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

media_muestral = 73.9615 # x
media_poblacional = 78.2676 # 
desviacion_muestral = 9.0361 # s
n = 100 # Tamao de la muestra
alpha = 0.05 # Nivel de significancia

t_calculado = (media_muestral - media_poblacional) / (desviacion_muestral / np.sqrt(n))
df = n - 1 # Grados de libertad
t_critico = stats.t.ppf(1 - alpha / 2, df)

x = np.linspace(-4, 4, 1000) # Rango de valores en el eje x
y = stats.t.pdf(x, df) # Funcin de densidad de probabilidad

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribucin t-Student", color='black')

x_rechazo_derecha = np.linspace(t_critico, 4, 100)
y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Regin de rechazo")

x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"t crtico = {t_critico:.2f}")
plt.axvline(x=-t_critico, color='black', linestyle='dashed')
plt.axvline(x=t_calculado, color='green', linestyle='dashed', label=f"t calculado = {t_calculado:.2f}")

plt.title("Distribucin t-Student y regiones de rechazo")
plt.xlabel("Valor de t")
plt.ylabel("Densidad de probabilidad")
plt.legend()

plt.grid()
plt.show()

print(f"Valor calculado de t: {t_calculado:.2f}")
print(f"Valor crtico de t (bilateral, ={alpha}): {t_critico:.3f}")

if abs(t_calculado) > t_critico:
  print("Rechazamos la hiptesis nula. Los tiempos de entrega de los lotes A y B no son iguales..")
else:
  print("No rechazamos la hiptesis nula. No hay evidencia suficiente para afirmar una diferencia.")
```
