```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos del problema
media_muestral = 76.96 # x̄
media_poblacional = 78.28 # μ
desviacion_muestral = 76.11 # s
n = 100 # Tamaño de la muestra
alpha = 0.05 # Nivel de significancia

# Paso 2: Cálculo del estadístico t
t_calculado = (media_muestral - media_poblacional) / (desviacion_muestral / np.sqrt(n))

# Paso 3: Determinación del valor crítico
df = n - 1 # Grados de libertad
t_critico = stats.t.ppf(1 - alpha / 2, df) # Valor crítico de t para prueba bilateral

# Paso 4: Gráfica de la distribución t-Student
x = np.linspace(-4, 4, 1000) # Rango de valores en el eje x
y = stats.t.pdf(x, df) # Función de densidad de probabilidad

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribución t-Student", color='blue')

# Sombrar la región de rechazo
x_rechazo_derecha = np.linspace(t_critico, 4, 100)
y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Región de rechazo")

x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

# Líneas de referencia
plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"t crítico = {t_critico:.2f}")
plt.axvline(x=-t_critico, color='black', linestyle='dashed')
plt.axvline(x=t_calculado, color='green', linestyle='dashed', label=f"t calculado = {t_calculado:.2f}")

plt.title("Distribución t-Student y regiones de rechazo")
plt.xlabel("Valor de t")
plt.ylabel("Densidad de probabilidad")
plt.legend()

plt.grid()
plt.show()

# Paso 5: Comparación y decisión
print(f"Valor calculado de t: {t_calculado:.2f}")
print(f"Valor crítico de t (bilateral, α={alpha}): ±{t_critico:.3f}")

if abs(t_calculado) > t_critico:
    print("Rechazamos la hipótesis nula. La media del grupo es significativamente diferente.")
else:
    print("No rechazamos la hipótesis nula. No hay evidencia suficiente para afirmar una diferencia.")
```
