```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n1 = 10
p1 = 0.05

prob_10 = binom.pmf(1, n1, p1)

# Parámetros del problema
n = 100 # Número de accidentes
p = prob_10 # Probabilidad de encestar una accidente

# Crear la distribución binomial
x = np.arange(0, n+1) # Valores posibles (0 a n)
pmf = binom.pmf(x, n, p) # Función de masa de probabilidad

# Gráfica de la distribución
plt.bar(x, pmf, color='green', edgecolor='black')
plt.title('Distribución Binomial: Accidentes de trancisto')


plt.xlabel('intersecciones de accidentes de trancito')
plt.ylabel('Probabilidad')
plt.show()

# Calcular probabilidades específicas
prob_0_successes = binom.pmf(0, n, p)

prob_3_successes = binom.pmf(3, n, p)


# Calcular valor esperado y varianza
expected_value = n * p
variance = n * p * (1 - p)

print(f"la probabilidad de 0 accidentes cada 10 minutos: " + str(prob_0_successes)  )
print(f"la probabilidad de 3 accidentes cada 10 minutos: " + str(prob_3_successes)  )
print(f"Valor esperado: {expected_value}")
print(f"Varianza: {variance}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_ = 75

x = np.arange(0, 100)
pmf = poisson.pmf(x, lambda_)

plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title('Distribución de Poisson: Carros que pasan por una intersección en 15 minutos')
plt.xlabel('Número de carros')
plt.ylabel('Probabilidad')
plt.show()

prob_75_carros = poisson.pmf(75, lambda_)

expected_value = lambda_
variance = lambda_

print(f"Probabilidad de que pasen 75 carros en 15 minutos por dicha intersección: {prob_75_carros:.4f}")
print(f"Número promedio de carros: {expected_value}")
print(f"Varianza: {variance}")
```
